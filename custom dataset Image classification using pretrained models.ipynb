```python
import os
import numpy as np

from tensorflow.keras.preprocessing.image import ImageDataGenerator
import pandas as pd
from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.layers import Dense,Input,Conv2D,MaxPooling2D,Flatten,Dropout
from tensorflow.keras.optimizers import SGD,Adam

from tensorflow.keras.applications import VGG19
```

```python
# ImageDataGenerator??
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
test_datagen = ImageDataGenerator(
    rescale=1./255
)
```

```python
train_data_gen = train_datagen.flow_from_directory(
                                                    directory="./data/PetImages_train_test/train",
                                                    target_size=(128, 128),
                                                    class_mode='binary'
                                                )
```

```python
test_data_gen = test_datagen.flow_from_directory(
                                                    directory="./data/PetImages_train_test/test",
                                                    target_size=(128, 128),
                                                    class_mode='binary'
                                                )
```

```python
base_model = VGG19(include_top=False,weights='imagenet',input_shape=train_data_gen.image_shape)
```

```python
model = Sequential()
model.add(base_model)
model.add(Flatten())

model.add(Dense(256,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
base_model = VGG19(include_top=False,weights='imagenet',input_shape=train_data_gen.image_shape)
base_model.trainable = False
model = Sequential()
model.add(base_model)
model.add(Flatten())

model.add(Dense(256,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation="sigmoid"))

model.summary()
```

```python
opt = SGD(learning_rate=0.01,momentum=0.9)
model.compile(optimizer=opt,loss="binary_crossentropy",metrics=["accuracy"])
history = model.fit(train_data_gen,epochs=10,batch_size=64)
model.save("pretrained_model_vgg19.keras")
```
