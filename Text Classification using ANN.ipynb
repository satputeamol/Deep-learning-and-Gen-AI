```python
df_train = pd.read_csv("./data/Covid/Corona_NLP_train.csv",encoding='latin-1')
```

```python
wl = WordNetLemmatizer()
stp_wrds = stopwords.words("english")
translator = str.maketrans('','',string.punctuation)
```

```python
def clean_text(text):
    word_tok = word_tokenize(text)
    lemma_word = [wl.lemmatize(word,pos='v') for word in word_tok if word.lower() not in stp_wrds]
    lowercase_lemma_word = [word.lower() for word in lemma_word]
    processed_text = " ".join(lowercase_lemma_word).translate(translator).strip()
    return processed_text
```

```python
text = "Hello, World! Python is amazing."
clean_text(text)
```

```python
df_train["Cleaned_tweet"] = df_train["OriginalTweet"].apply(lambda x:clean_text(x))
```

```python
df_train.head()
```

```python
X = df_train["Cleaned_tweet"].values
y = df_train["Sentiment"].values
```

```python
X[0]
```

```python
X.shape
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

```python
y_train = to_categorical(y, num_classes=5)
```

```python
y_train
```

```python
# tokenizer = Tokenizer(num_words=10,oov_token="<OOV>")
```

```python
tokenizer = Tokenizer(num_words=1000)
```

```python
tokenizer.fit_on_texts(X)
```

```python
word_index = tokenizer.word_index
```

```python
X[0]
```

```python
word_index
```

```python
sequences = tokenizer.texts_to_sequences(X)
```

```python
sequences
```

```python
X_train = pad_sequences(sequences, maxlen=50, padding='post')
```

```python
X_train[0].shape
```

```python
vocab_size = len(word_index)+1
embedding_dim = 16
max_len = 10
```

```python
vocab_size
```

```python
model = Sequential()
model.add(Input(shape = X_train[0].shape))
model.add(Embedding(input_dim = 1000,output_dim=embedding_dim))
model.add(GlobalAveragePooling1D())
model.add(Dense(32,activation="relu"))
model.add(Dense(16,activation="relu"))
model.add(Dense(5,activation="softmax"))
```

```python
model.summary()
```

```python
16000/17157
```

```python
model.compile(optimizer="sgd",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train,y_train,epochs=20,verbose=1)
```
