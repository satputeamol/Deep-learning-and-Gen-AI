```python
import chromadb
```

```python
chroma_client = chromadb.Client()
```

```python
collections = chroma_client.create_collection("chroma-db-lesson")
```

```python
collections.add(
    ids = ['id1','id2'],
    metadatas = [{"source":"geography"},{"source":"food"}],
    documents = ["The capital of India is New Delhi","Banana is a fruit"]
)
```

```python
results = collections.query(
                                query_texts = ["What is the capital of India"],
                                n_results = 1
                        )
```
