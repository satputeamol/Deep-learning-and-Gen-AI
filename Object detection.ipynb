```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

```python
model = YOLO(model='yolo11n.pt')
```

```python
filepath = "./Object_detection/image_for_object_detection.jpeg"
```

```python
img = cv2.imread(filepath)
```

```python
plt.imshow(img)
plt.show()
```

```python
res = model(img)
```

```python
res
```

```python
annotated_img = res[0].plot()
```

```python
plt.imshow(annotated_img)
plt.show()
```

```python
res = model.predict(img,conf=0.5)
```

```python
annotated_img = res[0].plot()
```

```python
plt.imshow(annotated_img)
plt.show()
```

```python
output_filepath = "./Object_detection/object_detection_output.jpeg"
cv2.imwrite(output_filepath,annotated_img)
```
