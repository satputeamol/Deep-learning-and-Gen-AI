## POS :: part of speech tagging

```python
## POS :: part of speech tagging 

# Common Parts of Speech Tags:
# 	•	Noun (NN): Person, place, or thing (e.g., cat, happiness)
# 	•	Pronoun (PRP): He, she, it
# 	•	Verb (VB): Actions or states (e.g., run, is)
# 	•	Adjective (JJ): Describes nouns (e.g., beautiful, large)
# 	•	Adverb (RB): Describes verbs or adjectives (e.g., quickly, very)
# 	•	Preposition (IN): Shows relationship (e.g., on, at, in)
# 	•	Conjunction (CC): Connects words or sentences (e.g., and, but)
# 	•	Determiner (DT): Specifies a noun (e.g., the, some)
# 	•	Interjection (UH): Expresses emotion (e.g., wow, oh)
# 	•	Proper Noun (NNP): Names (e.g., John, India)
```

## Named entity recognition

```python
# Common NER Categories:
# 	•	Person (PER): Names of people (e.g., Elon Musk, Marie Curie)
# 	•	Organization (ORG): Companies, institutions, or agencies (e.g., Google, United Nations)
# 	•	Location (LOC): Cities, countries, or geographical landmarks (e.g., Paris, Mount Everest)
# 	•	Date (DATE): Specific dates or periods (e.g., January 2023, 21st century)
# 	•	Time (TIME): Time expressions (e.g., 2:00 PM, morning)
# 	•	Money (MONEY): Monetary values (e.g., $500, €10 million)
# 	•	Percent (PERCENT): Percentages (e.g., 20%, 85%)
# 	•	Event (EVENT): Named events (e.g., Olympics, World War II)
# 	•	Product (PRODUCT): Items or brands (e.g., iPhone, Tesla Model S)
# 	•	Miscellaneous (MISC): Entities that don’t fit standard categories
```

#### BOW

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
import pandas as pd
```

```python
documents = ["This is NLP NLP session","nlp uses machine learning","NLP uses deep deep learning"]
```

```python
cv = CountVectorizer(stop_words="english")
```

```python
cv.fit(documents)
```

```python
cv.vocabulary_
```

```python
cv.vocabulary
```

```python
vocab = sorted(cv.vocabulary_.keys())
```

```python
doc_vector = cv.transform(documents).toarray()
```

```python
doc_vector
```

```python
df = pd.DataFrame(doc_vector,columns = vocab)
```

```python
df
```

#### TF-IDF

```python
tfidf = TfidfVectorizer(stop_words="english")
```

```python
tfidf.fit(documents)
```

```python
vocab = sorted(tfidf.vocabulary_.keys())
```

```python
doc_vector = tfidf.transform(documents).toarray()
```

```python
df = pd.DataFrame(doc_vector,columns = vocab)
```

#### Embeddings

```python
## Embeddings :: Embeddings in NLP refer to a dense vector representation of words or other textual units (like sentences or documents) 
## in a continuous vector space, where semantically similar words are mapped closer together.
# Popular Embedding Methods
# 	1.	Word2Vec (Google)
# 	•	Trained using Continuous Bag of Words (CBOW) or Skip-gram models.
# 	•	Example: “king - man + woman ≈ queen”
# 	2.	GloVe (Global Vectors by Stanford)
# 	•	Trained on word co-occurrence statistics from a large corpus.
# 	•	Captures both local and global statistical relationships.
# 	3.	FastText (Facebook)
# 	•	Extends Word2Vec by using character n-grams, making it better for rare and morphologically complex words.
# 	4.	BERT (Bidirectional Encoder Representations from Transformers by Google)
# 	•	Contextual embeddings based on transformer architecture.
# 	•	“Apple” can have different embeddings depending on whether it refers to the fruit or the company.
# 	5.	GPT Embeddings (OpenAI)
# 	•	Contextual embeddings useful for complex text understanding and generation tasks.
```
