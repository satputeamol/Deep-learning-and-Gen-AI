```python
import numpy as np
from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.layers import Dense,LSTM,RNN,Bidirectional,Input,SimpleRNN
from tensorflow.keras.optimizers import Adam,SGD
```

```python
step_size = 3
X = []
y = []

for i in range(len(seq)):
    if step_size+i < len(seq):
        X.append(seq[i:step_size+i])
        y.append(seq[step_size+i])
```

```python
x_arr = np.array(X)
```

```python
X_reshaped = x_arr.reshape(x_arr.shape[0],x_arr.shape[1],1)
```

```python
SimpleRNN??
```

```python
model = Sequential()
model.add(Input(shape=X_reshaped[0].shape))
model.add(SimpleRNN(units=256,activation='tanh',return_sequences=True,unroll=True))
model.add(Dense(16,activation='relu'))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer="sgd",loss="mse",metrics=["accuracy"])
```

```python
model.fit(X_reshaped,y_arr,epochs=10,verbose=1)
```
