```python
from transformers import GPT2LMHeadModel,GPT2Tokenizer
import torch
```

```python
pretrained_model = 'gpt2'

model = GPT2LMHeadModel.from_pretrained(pretrained_model)
tokenizer = GPT2Tokenizer.from_pretrained(pretrained_model)
```

```python
text = "The future of machine learning is"
max_length = 50
```

```python
tokenizer.encode(text,return_tensors='pt')
```

```python
tokenizer.encode(text,return_tensors='tf')
```

```python
tokenizer.encode(text,return_tensors='np')
```

```python
inputs = tokenizer.encode(text,return_tensors='pt')
```

```python
def text_generation(inputs,max_length,ret_seq):
    with torch.no_grad():
        outputs = model.generate(
            inputs = inputs,
            max_length = max_length,
            num_return_sequences=ret_seq,
            no_repeat_ngram_size=2,
            top_k = 50,
            top_p = 0.95,
            do_sample=True
        )
    generated_text = [tokenizer.decode(output) for output in outputs]
    return generated_text
```

```python
def text_generation(inputs,max_length,ret_seq,top_k,top_p):
    with torch.no_grad():
        outputs = model.generate(
            inputs = inputs,
            max_length = max_length,
            num_return_sequences=ret_seq,
            no_repeat_ngram_size=2,
            top_k = top_k,
            top_p = top_p,
            do_sample=True
        )
    generated_text = [tokenizer.decode(output) for output in outputs]
    return generated_text
```
